```
import dimod
import networkx as nx
import matplotlib.pyplot as plot
```

```
def max_cut(G):
    
    """
    
    """
    
    if not G:
        G = random_graph(12)
        
    # draw network
    nx.draw(G, with_labels=True)
        
    # assign bias and coupling for graph
    h = {v: 0 for v in G}
    
    J = {(u, v): 1 for u, v in G.edges}
    
    # create a sampler
    ising_sampler = dimod.SimulatedAnnealingSampler()
    
    response = ising_sampler.sample_ising(h, J)
    sample = next(iter(response))
    
    
    return set(v for v in G if sample[v] >= 1)
```

```
def random_graph(node_size, edge_prob=0.5):
    
    """
    
    """
    
    return nx.gnp_random_graph(node_size, edge_prob)
```

```
# generate a random graph using networkx
G = random_graph(24)

print(max_cut(G))
```
